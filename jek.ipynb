```python
import pandas as pd
import altair as alt

url = "https://raw.githubusercontent.com/UIUC-iSchool-DataViz/is445_data/main/ufo-scrubbed-geocoded-time-standardized-00.csv"

df = pd.read_csv(url, header=None)

df.columns = [
    "datetime", "city", "state", "country", "shape",
    "duration_seconds", "duration", "comments",
    "date_posted", "latitude", "longitude"
]
df
```

```python
df['datetime'] = pd.to_datetime(df['datetime'], errors='coerce')
df['year'] = df['datetime'].dt.year
```

```python
import altair as alt
alt.data_transformers.disable_max_rows()

chart1 = alt.Chart(df).mark_line().encode(
    x=alt.X('year:O', title='Year'),
    y=alt.Y('count()', title='Number of Sightings')
).properties(
    title='UFO Sightings Over Time'
)

chart1
```

This visualization shows the number of UFO sightings over time using a line chart. The x-axis represents the year, which was derived from the original datetime variable, and the y-axis shows the total count of sightings. A line chart was chosen because it effectively illustrates trends and changes over time. The data was transformed by converting the datetime column into a year variable to simplify the time scale and make patterns easier to interpret. This visualization highlights a noticeable increase in reported sightings in more recent years.

```python
shapes = list(df['shape'].dropna().unique())
shapes.insert(0, "All")  # add "All" at the top

dropdown = alt.binding_select(options=shapes, name="Select Shape:")
selection = alt.selection_point(fields=['shape'], bind=dropdown)

chart2 = alt.Chart(df).mark_bar().encode(
    x=alt.X('shape:N', title='UFO Shape'),
    y=alt.Y('count()', title='Number of Sightings'),
    color='shape:N'
).transform_filter(
    (selection) | (alt.datum.shape == "All")
).add_params(
    selection
).properties(
    title='UFO Sightings by Shape (Interactive)'
)

chart2
```

This visualization displays the number of UFO sightings by shape using a bar chart. The x-axis represents different UFO shapes, while the y-axis shows the count of sightings for each shape. Color encoding was used to distinguish between the different categories of shapes, making the chart easier to read. A bar chart was selected because it allows for clear comparison between categorical values. The data required minimal transformation aside from cleaning missing values in the shape column.

```python
chart1.save('chart1.html')
chart2.save('chart2.html')
```
